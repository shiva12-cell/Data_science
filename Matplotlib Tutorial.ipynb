```python
#Histogram Tutorial
#with histogram only need single dimen00000000000tion of array and frequency
#Bin or bucket concept : really arranged
#lets play with bins parameter
```

```python
blood_sugar_women=[113,85,150,88,93,115,135,80,77,82,129]
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,210,250]
plt.xlabel("Sugar")
plt.ylabel("Total no of patients")
plt.title("Blood sugar Analysis")
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=.95,color=['green','orange']
           ,label=['men','women'])
plt.legend()
```

```python
blood_sugar_women=[113,85,150,88,93,115,135,80,77,82,129]
blood_sugar_men=[113,85,90,150,149,88,93,115,135,80,77,82,100]
plt.xlabel("Sugar")
plt.ylabel("Total no of patients")
plt.title("Blood sugar Analysis")
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,150],rwidth=.95,color=['green','orange']
           ,label=['men','women'],orientation="horizontal")
plt.legend()
```

```python
exp_vals = [1200, 150, 300, 60, 100, 80, 120, 50]
exp_labels = ["Rent", "Utilities", "Groceries", "Internet", "Entertainment", 
              "Transportation", "Insurance", "Miscellaneous"]
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.1f%%',shadow=True,explode=[0,0.2,0.2,0,0.2,0,0,0],startangle=180)
plt.axis("equal")# makes the circle properly equals
#autoct :decimal number 0.2 shows upto 2 decimal numbers
#radius :  makes size of pie chart
# explode : putout the pie 0.1 0.2 are level of expulsion
#startable : tilt / rotate the pie chart
```
